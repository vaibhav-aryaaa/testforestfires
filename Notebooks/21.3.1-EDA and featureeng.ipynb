### Data Cleaning - EDA

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

the dataset is converted into two sets based on region from 122th index, we can make a new column based on the region
1. "Bejia Region Dataset" - 0
2. "Sidi-Bel Abbes Region Dataset" - 1

add a new column with region

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df=dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Region']]=df[['Region']].astype(int)
df.head()
```

```python
df.isnull().sum()
```

dropping one record 122
1. remove the null values

```python
df=df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

fix spaces in the columns

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

change the required columns as integer datatype (from day to dw)

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

change other columns to float

```python
[features for features in df.columns if df[features].dtype=='O']
```

```python
objects=[features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i!= 'Classes': #'Classes' is classification variable
        df[i]=df[i].astype(float)
```

```python
df.info()
```

Monthly fire analysis

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
dftemp=df.loc[df['Region']==1]
# plt.subplot(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Month',weight='bold')
plt.title('Monthly Analysis Fire of Sidi-Bel Region',weight='bold')
plt.show()
```

```python
dftemp=df.loc[df['Region']==0]
# plt.subplot(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Month',weight='bold')
plt.title('Monthly Analysis Fire of Bejaia Region',weight='bold')
plt.show()
```

Its observed that August and September had the most number of forest fires for both regions. And from the above plot of months, we can understand few things. Most of the fires happened in August and very high Fires happened in only 3 months - June, July and August.  
Less Fires was on September
